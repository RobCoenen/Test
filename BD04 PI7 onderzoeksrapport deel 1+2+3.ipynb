**Wat is data cleaning?** <br>
Los de inconsistenties in de gegevens op en behandel de ontbrekende waarden.

**Data cleaning in dit project**
Zoals eerder omschreven hebben wij de dataset eerst verkend in Microsoft Excel en Power BI. Hierbij vielen ons een aantal zaken op, die we in dit hoofdstuk gaan verbeteren. Zie onderstaande code.

Op dit moment bevat de 'CarName' van de dataset zowel de fabrikant van de auto als het model, als voorbeeld: alfa-romero giulia. Wij willen dit graag splitsen, zodat de autofabrikant los komt te staan van de modelnaam.

```python
# Het splitsen van 'CompanyName' vanaf de 'CarName' kolom

# Deze code omschrijft dat er een variabele aangemaakt wordt met de naam 'CompanyName' en dat er in de cars variabele (het dataframe) gezocht moet worden naam de kolomnaam 'CarName'. 
# Vervolgens zorgt lambda er voor dat de gehele rij (x geeft het aantal rijen aan, waarbij x onbeperkt is, afgezocht moeten worden. 
# De ' ' geeft aan dat de regel afgebroken wordt na een spatie. 
# De 0 geeft tenslotte aan dat er op index 0 begonnen wordt, met andere woorden, vanaf het begin.)
# https://www.w3schools.com/python/python_lambda.asp
CompanyName = cars['CarName'].apply(lambda x : x.split(' ')[0]) 

# Hier wordt een kolom aangemaakt, op positie 3, met de naam 'CompanyName. Vervolgens worden de waarden (zoals zojuist "berekend" toegevoegd.)
# https://docs.python.org/3/tutorial/datastructures.html
cars.insert(3,"CompanyName",CompanyName)

# Hier wordt de bestaande tabel 'CarName' verwijderd.
# https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.drop.html
cars.drop(['CarName'],axis=1,inplace=True)

# We printen nu de output van deze bewerkingen
cars.head()
```

Verder is de 'car_ID' kolom niet erg interessant voor het werken met deze dataset. Deze verwijderen we.

```python
cars.drop(['car_ID'],axis=1,inplace=True)
```

Vervolgens kunnen we met de volgende code weergeven welke (unieke) waarden er nog zijn in de 'CompanyName' tabel.

```python
# https://numpy.org/doc/stable/reference/generated/numpy.unique.html
cars.CompanyName.unique()
```

Zoals te zien is zijn ook hier nog een aantal fouten in. Zo zijn er twee variabelen namelijk 'maxda' en 'mazda', is 'Porsche' eenmalig verkeerd omgeschreven ('porcshe') evenals dat 'Toyota' ook als 'Toyouta' omschreven wordt en voor 'Volkswagen' soms de afkoring 'vw' gebruikt wordt. Deze spelfouten gaan we verbeteren

```python
# Zoekt in de 'cars' dataframe naar de kolomnaam 'CompanyName' en zorgt dat alle records zonder hoofdletters worden.
# https://www.programiz.com/python-programming/methods/string/lower
cars.CompanyName = cars.CompanyName.str.lower()

# Deze functie maakt twee variabelen aan waarbij A de bestaande naam is en B de gewijzigde naam. Inplace maakt het mogelijk dat de waarden overschreven worden.
# https://www.rdocumentation.org/packages/textclean/versions/0.9.3/topics/replace_names
def replace_name(a,b):
    cars.CompanyName.replace(a,b,inplace=True)

# Hier worden de waarden ingegeven die vervangen worden, voor de komma de A waarde en na de komma de B waarde.
replace_name('maxda','mazda')
replace_name('porcshce','porsche')
replace_name('toyouta','toyota')
replace_name('vw','volkswagen')
replace_name('vokswagen','volkswagen')

# Alternatieve code
# data = data.replace(to_replace ="maxda", value ="mazda") 
# data = data.replace(to_replace ="porcshce", value ="porsche") 
# data = data.replace(to_replace ="toyouta", value ="toyota") 
# data = data.replace(to_replace ="vw", value ="volkswagen")
# data = data.replace(to_replace ="vokswagen", value ="volkswagen") 
# data = data.replace(to_replace ="Nissan", value ="nissan")

#Vervolgens printen we nogmaals de waarden uit de 'CompanyName' tabel.
cars.CompanyName.unique()
```

```python
upper=df_corr.where(np.triu(np.ones(df_corr.shape),k=1).astype(np.bool))
```

### Multiple Linear Regression

<strong>Algemene omschrijving:</strong><br>
Multiple Linear Regression valt onder de vorm 'Supervised Learning' (begeleid leren) van machine learning. Dat wil zeggen dat je data verzameld/produceert op basis van historische gegevens. Er is een reeks data waarop het model getraind kan worden en op basis van deze trainingsdata is het mogelijk relaties af te leiden tussen de (ofhankelijke) variabelen, daardoor is mogelijk een voorspelling te maken; <br>

• Regressie is het voorspellen van een (nummerieke) waarde; <br>
• Regressieanalyse is een techniek die gebruikt wordt om modellen te voorspellen; <br>
• In regressieanalyse wordt de relatie onderzocht tussen een afhankelijke (doel) en een onafhankelijke variabele(n) (de voorspeller); <br>
• Het helpt om te begrijpen hoe de waarde van de afhankelijke variabele verandert wanneer één van de onafhankelijke variabelen wordt gevarieerd terwijl andere, onafankelijke, variabelen onveranderd blijven; <br>
• In alle verschillende gevallen van regressieanalyse moet een regressiefunctie, die een functie is van de onafhankelijke variabelen, worden geschat.

De doelen van regressieanalyse zijn: <br>
• De (significante) relaties aan geven tussen afhankelijke en onafhankelijke variabelen; <br>
• De sterke aangeven van de invloed van meerdere onafhankelijke variabelen op een afhankelijke variabele. <br>

Een regressiemodel bestaat uit de volgende parameters: <br>
• De onbekende parameters (𝛽_𝑖) <br>
• De onafhankelijke variabelen (𝑥_𝑖) <br>
• De afhankelijke variabele (𝑦) <br>

Er zijn meerdere vormen van regressie, te weten linear, logistic en polynomial. 
In deze uitwerking wordt de lineare regressie behandeld. <br>

Lineare regressie vindt een verband tussen afhankelijke en een of meer onafhankelijke variabelen door een rechte lijn te passen (de best passende lijn) te realiseren.

### Neurale Netwerken

---

Een neuraal netwerk bestaat uit een input-laag, hidden layers en een output laag. <br>

• Een neuron is een knooppunt binnen een neuraal netwerk welke meerdere input-waarden kan hebben en welke één output waarde vormt. Deze wordt berekend door de activatiefunctie (wiskundig gegeven die niet-lineariteit introduceren binnen netwerken). <br>
• Gewichten bepalen hoe zwaar een functie meetelt in het trainen van het model. Bij gewicht 0 telt de bijbehorende functie niets bij. <br>
• Een neuraal netwerk maakt het mogelijk om niet-lineaire output te voorspellen. <br>
• Gradient Descent / Backpropagation zijn optimizers. Deze leren iedere keer als input/output door het neuraal netwerk is gegaan. <br>
• Optimizer bepaalt de gewichten en biases om zo verlies van info te verminderen. <br>
• Weights zijn waarden die vermeningvuldigd worden door de input features om zo een output waarde voorspellen. Biases zijn de waarde van een output met het gewicht 0. <br>
• Convergeren betekent dat op ten duur een algoritme een optimaal antwoord bereikt. <br>
• De learning rate bepaalt hoe de optimizers de gewichten en biases veranderen. High learning rate traint snellers maar bereikt mogelijk zijn nooit zijn optimale antwoord maar een lower rate traint langzamer. <br>
• Dropout: verwijdert een random selectie van units in een netwerk laag voor één doorloop van het neuraal netwerk. Early Stopping: methode voor regularisatie die vroegtijdig wordt beëindigd <br>
• Gradient-descent: techniek om verlies te minimaliseren door met in acht houding van de parameters en gecontioneerd op de trainingsdata. <br>
• EPOCH: Een volledig doorloop naar voren en naar achteren binnen een neuraal netwerk. <br>
• Batch size: Totaal aantal trainingsvoorbeelden aanwezig in een enkele batch <br>
• Iteraties is het aantal batches dat nodig is om één epoch te voltooien. We kunnen de dataset van 2000 voorbeelden in batches van 500 verdelen, dan duurt het 4 iteraties om 1 EPOCH te voltooien

### Support Vector Machines

---

• Een support vector machine is ontworpen voor binaire classificatie (dus 2 mogelijke uitkomsten zoals ja/nee, kop of munt); <br>
• Je kunt SVM gebruiken ook gebruiken om numerieke waarden te voorspellen (regressie); <br>
• Bij SVM voorspel je zaken door data in groepen te verdelen; <br>
• Het idee: splits de examples (waarnemingen) in twee groepen en scheid ze door een hyperplane. <br>

• Support Vectors zijn de punten van verschillende groepen waar als je hier parallelle lijnen door zou trekken, de onderlinge afstand tussen deze lijnen (2 dimensies) het grootst is; <br>
• Het hyperplane is de lijn (of het vlak) wat hier tussenin ligt, oftewel de scheidingslijn (de maximum margin seperator/classifier); <br>
• De afstand tussen de stippellijnen noemen we de marge (margin). <br>
• Alle andere punten (buiten de support vectors) kunnen vrij gewijzigd worden zonder dat dit de seperator beïnvloed. Dit voorkomt overfitting. <br>
